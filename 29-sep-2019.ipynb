#### There are two types of functions for taking values from csv data

- iloc (based on position)
- loc(based on label)

```python
o.loc[2,'genre']
```

```python
o.iloc[2,5]
```

```python
o.iloc[1,:]
```

```python
o.iloc[:,2:]
```

```python
o.iloc[:,3:].min()
```

```python
o.iloc[:,3:].max()
```

```python
o.loc[(o['duration']>150) | (o['duration']<170)]
```

```python
o.loc[(o['duration']>150) & (o['duration']<170)]
```
